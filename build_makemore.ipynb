## Counting bigrams in a 2D torch tensor

```python
import torch
import numpy
```

```python
# As an example, create a torch tensor as 3x5 matrix of zeros with integer type for storing counts
a = torch.zeros((3,5), dtype = torch.int32)

# print the array of zeros
print (a, '\n')
print

# set the element at row 0, column 0 to 5 and print the array
a[0,0] = 5
print ("a[0,0] = 5 results in \n", a)

# set the element at row 1, column 3 to 4 and print the array
a[1,3] += 4
print ("\na[1,3] += 4 results in \n", a)
```

```python
# Define a 2D torch tensor of 28x28 zeros (26 letters + start & end characters) with integer type for storing counts
N = torch.zeros((28,28), dtype=torch.int32)

# each row corresponds to the first character of the bigram
# each column corresponds to the second character of the bigram
# the value in each cell is the count of how many times the second character follows the first character
print (N)
```

```python
# Prepare to create a look up table for converting characters to indices
# Create a set of all characters in the list of words
s = set(''.join(words))
# sort the set of characters
chars = sorted (s)
# print the length and sorted list of characters
print (len(s), "letters sorted:", chars)
```

```python
# Create a dictionary for look up into the bigram array with key:value pairs = character:integer_index
stoi = {ch:i for i,ch in enumerate(chars)}

# Add special start <S> and end <E> characters and their indices
stoi['<S>'] = 26
stoi['<E>'] = 27
stoi
```

```python
# Define a 2D torch tensor of 28x28 zeros (26 letters + start & end characters) with integer type for storing counts
N = torch.zeros((28,28), dtype=torch.int32)

# Fill it for just the first word in the list of words, "Emma"

# for each word in the list of words
for w in words[:1]:
    # get the list of characters in the word
    chs = ['<S>'] + list(w) + ['<E>']
    # for each pair of characters in the word
    for ch1, ch2 in zip(chs, chs[1:]):
        # get the index of the first character
        iX1 = stoi[ch1]
        # get the index of the second character
        iX2 = stoi[ch2]
        # increment the cell value = count of the bigram for the pair of characters
        N[iX1, iX2] += 1
        # print the characters in each pair (bigram) and their indices
        print (ch1, ":", iX1, ",", ch2, ":", iX2)
        # print the value of the cell for the bigram
        print ("N[{}][{}]=".format (str(iX1), str(iX2)), N[iX1, iX2].item(), "\n")
```

```python
# What does our 28X28 matrix, N, look like after processing the first word, "Emma"?
# Note the five 1s in [First char][Second char] cells corresponding to "<S>e", "em", "mm", "ma", and "a<E>".
print (N)
```

```python
# Reinitialize N: 26 letters + <S> + <E> = 28X 28 matrix
N = torch.zeros((28,28), dtype=torch.int32)

# Let's process all the words now
for w in words:
    chs = ['<S>'] + list(w) + ['<E>']
    for ch1, ch2 in zip(chs, chs[1:]):
        iX1 = stoi[ch1]
        iX2 = stoi[ch2]
        N[iX1, iX2] += 1

# What does our 28X28 matrix, N, look like after processing all the words?
print (N)

# Recall these are the bigrams we found earlier and their counts in the corpus
# Look for these numbers in the output matrix in the corresponding [row][col] cells
# [(('n', '<E>'), 6763),
# (('a', '<E>'), 6640),
# (('a', 'n'), 5438),
# (('<S>', 'a'), 4410),
# (('e', '<E>'), 3983),
# (('a', 'r'), 3264),
# (('e', 'l'), 3248),
# (('r', 'i'), 3033),
# (('n', 'a'), 2977),
# (('<S>', 'k'), 2963),
# (('l', 'e'), 2921),
#...
#...
```

```python
# The previous visualization is a good first step, but not very useful
# Let's try a better visualization
# Let's invert stoi to get itos = integer to character look up
itos = {ch:i for ch, i in enumerate(chars)}
itos
```

```python
# we define a 27x27 matrix of zeros not 28x28 because we don't want to plot the <S> and <E> characters
# Instead we will replace both with a . (dot) character
# So now the matrix is 27x27 = 26x26 letters + . (dot) character's row and column
N = torch.zeros((27, 27), dtype=torch.int32)
```

```python
# Get sorted list of characters from the set of characters in the list of words
chars = sorted(list(set(''.join(words))))
# Start with the second index = 1 because we will reserve index 0 for the dot character
# Create a dictionary for look up into the bigram array with key:value pairs = character:integer_index
stoi = {s:i+1 for i,s in enumerate(chars)}
# Add the dot character and its index = 0
stoi['.'] = 0
# Inverse the dictionary to get integer:index look up
itos = {i:s for s,i in stoi.items()}

# Fill the matrix as before
for w in words:
  chs = ['.'] + list(w) + ['.']
  for ch1, ch2 in zip(chs, chs[1:]):
    ix1 = stoi[ch1]
    ix2 = stoi[ch2]
    N[ix1, ix2] += 1

# print the matrix
print (N)
```

```python
# A better visualization of the matrix
import matplotlib.pyplot as plt
%matplotlib inline

# 16x16 figure size
plt.figure(figsize=(16,16))
# colorscale = yellow green
plt.imshow(N, cmap='YlGn')

# for each row in the matrix
for i in range(27):
    # for each column in the matrix
    for j in range(27):
        # concatenate the characters (into a string) corresponding to the row and column indices
        chstr = itos[i] + itos[j]
        # print the string in the middle of the corresponding cell
        plt.text(j, i, chstr, ha="center", va="bottom", color='gray')
        # print the bigram count = value of the cell in the middle of the corresponding cell
        plt.text(j, i, N[i, j].item(), ha="center", va="top", color='gray')
# turn off the axis
plt.axis('off');
```

## Sampling from the model
The next step is to sample based on this 2D bigram tensor.

### Example of probablity distribution and sampling

```python
# List the first row of the matrix as denoted by the row index = 0
p = N[0].float()
print (p, '\n')

# Print the sum of the counts in the row
s = p.sum().item()
print (s)

# Get probabilities by dividing each count by the sum
p = p/s
p
```

```python
# Use the torch generator to generate random numbers with seed = 2147483647
g = torch.Generator().manual_seed(2147483647)

# Generate a random tensor of 3 elements
p = torch.rand(3, generator=g)
print (p)

# probability distribution of the three elements in p
p = p/p.sum()
print (p)
```

```python
# Generate a sample of 100 elements from the probability distribution in p with replacement
# We use replacement because we want to sample the same element more than once
# Using torch.multinomial() for this
torch.multinomial(p, num_samples=100, replacement=True, generator=g)
# Note that the counts of the three elements (0,1,2) in the sample are roughly proportional to their probabilities in the previous cell output
```

### Real example that steps through for one predicted name (word)

```python
g = torch.Generator().manual_seed(2147483647)

# how many names to generate
for i in range(1):
    # start with an empty list
    out = []
    # index of the first character
    ix=0

    # keep generating characters until we hit the end-of-word character
    while True:
        p = N[ix].float()
        print ("\nBigram counts for the sampled row index=", ix, "are:\n", p)
        # probability distribution of the characters
        p = p/p.sum()
        print ("Probablity distribution for the row index=", ix, "are:\n", p)
        # Generate a sampled index based on the probability distribution
        ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
        # print the sampled index and corresponding character
        print ("Sampled index", ix, "character:", itos[ix])
        # append the character to the output list
        out.append(itos[ix])
        print ("In progress string:", ''.join(out))
        if (ix==0):
            break

    # print the word
    print ("\nFinal string:", ''.join(out))
```

## Loss function and logprobs

```python
# GOAL: maximize likelihood of the data w.r.t. model parameters (statistical modeling)
# equivalent to maximizing the log likelihood (because log is monotonic)
# equivalent to minimizing the negative log likelihood (NLL)

# log(a*b*c) = log(a) + log(b) + log(c)
# is the same as loss function = loss1 + loss2 + loss3
```

```python
log_likelihood = 0
n = 0

# for each word in the first 3 words only
for w in words[:3]:
  chs = ['.'] + list(w) + ['.']
  for ch1, ch2 in zip(chs, chs[1:]):
    ix1 = stoi[ch1]
    ix2 = stoi[ch2]
    # probability of the bigram
    prob = P[ix1, ix2]
    # log of the probability
    logprob = torch.log(prob)
    # add the log probability to the running total
    log_likelihood += logprob
    print (F'{ch1}{ch2}:{prob:.4f} {logprob:.4f}')
    n += 1

print (f'{log_likelihood=}')

# negative log likelihood
nll = -log_likelihood
print (f'{nll=}')
# average negative log likelihood
print (f'{nll/n:.4f}')
```

```python
log_likelihood = 0
n = 0

# for each word in the entire corpus
for w in words:
  chs = ['.'] + list(w) + ['.']
  for ch1, ch2 in zip(chs, chs[1:]):
    ix1 = stoi[ch1]
    ix2 = stoi[ch2]
    # probability of the bigram
    prob = P[ix1, ix2]
    # log of the probability
    logprob = torch.log(prob)
    # add the log probability to the running total
    log_likelihood += logprob
    # print (F'{ch1}{ch2}:{prob:.4f} {logprob:.4f}')
    n += 1

print (f'{n=}')
print (f'{log_likelihood=}')

# negative log likelihood
nll = -log_likelihood
print (f'{nll=}')
# average negative log likelihood
print (f'{nll/n:.4f}')
```

```python
# Create the training set of bigrams with xs and ys as the input and target indices
xs, ys = [], []

# for each word in the entire corpus
for w in words[:1]:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
            ix1 = stoi[ch1]
            ix2 = stoi[ch2]
            print(ch1, ch2)
            xs.append(ix1)
            ys.append(ix2)
xs = torch.tensor(xs)
print (xs)

ys = torch.tensor(ys)
print (ys)
```

```python
# Create the training set of bigrams by converting xs and ys to one-hot vectors
# one hot encoding is a way to represent categorical data in a vector format
import torch.nn.functional as F

# one hot encoding of xs with 27 classes explicitly specified
xenc = F.one_hot(xs, num_classes=27)

# the rows are the samples and the columns are like the features
# for each sample, there are 27 features, one for each class
# for the five samples, the shape is (5, 27)
print (xenc.shape)
# In each row, there is a "1" at the index represented by the value in xs
print (xenc)
plt.imshow(xenc)
```
